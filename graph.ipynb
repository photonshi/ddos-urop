```python
import pandas as pd
from matplotlib import pyplot
```

```python
loss = pd.read_csv("loss.csv")
no_loss = pd.read_csv("noloss.csv")

def get_diff(l:str) -> list:
    '''
    helper function that extracts difference between http pairs and return output as list
    '''
    # for loop to get every other difference pair
    out = list()
    for i in range(0, len(l)-2, 2):
        curr_t, next_t = l["_ws.col.Time"].iat[i], l["_ws.col.Time"].iat[i+1]
        out.append(next_t - curr_t)
    return out
```

now we go through each column and populate a new dataframe with only rtt

```python
loss_http = loss[loss["_ws.col.Protocol"] == "HTTP"]
noloss_http = no_loss[no_loss["_ws.col.Protocol"] == "HTTP"]

rtt = pd.DataFrame()

rtt["loss"] = get_diff(loss_http)
rtt["no_loss"] = get_diff(noloss_http)

# make dataframe for plotting
p = pd.melt(rtt, value_vars=["loss", "no_loss"])
```
